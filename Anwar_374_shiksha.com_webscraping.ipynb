```python
import re
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
# Setup Selenium (headless)
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(service=Service(), options=options)
```

```python
# Prepare page URL
base_url = "https://www.shiksha.com/engineering/ranking/top-engineering-colleges-in-india/44-2-0-0-0?pageNo=8"
```

```python
# Define empty lists
NIRF_Rank = []
Institute = []
Placements_rating = []
Infrastructure_rating = []
Faculty_Course_Curriculum_rating = []
Crowd_Campus_Life_rating = []
Value_for_Money_rating = []
Learning_Experience_rating = []
Total_Reviews = []
Fees = []
Salary = []

# Loop through 5 pages (you can increase)
for page in range(1):
    print(f"Scraping page {page}...")
    driver.get(base_url.format(page))
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, "html.parser")

    for i in soup.find_all("section", class_="rank_tuplev1 bg-white clear_float ranking-flex"):   # Container
        text = i.text

        nirf = re.findall(r'(\d+)NIRF', text)
        if nirf:
            NIRF_Rank.append(nirf[0])
        else:
            NIRF_Rank.append(np.nan)

        name = re.findall(r"NIRF '\s*\d+(.*?)\d+Placements", text)
        if name:
            Institute.append(name[0].strip())
        else:
            Institute.append(np.nan)

        place = re.findall(r'Placements(\d\.\d)', text)
        if place:
            Placements_rating.append(place[0])
        else:
            Placements_rating.append(np.nan)

        infra = re.findall(r'Infrastructure(\d\.\d)', text)
        if infra:
            Infrastructure_rating.append(infra[0])
        else:
            Infrastructure_rating.append(np.nan)

        faculty = re.findall(r'Faculty.*?(\d\.\d)', text)
        if faculty:
            Faculty_Course_Curriculum_rating.append(faculty[0])
        else:
            Faculty_Course_Curriculum_rating.append(np.nan)

        crowd = re.findall(r'Crowd.*?(\d\.\d)', text)
        if crowd:
            Crowd_Campus_Life_rating.append(crowd[0])
        else:
            Crowd_Campus_Life_rating.append(np.nan)

        value = re.findall(r'Value for Money(\d\.\d)', text)
        if value:
            Value_for_Money_rating.append(value[0])
        else:
            Value_for_Money_rating.append(np.nan)

        learn = re.findall(r'Learning Experience(\d\.\d)', text)
        if learn:
            Learning_Experience_rating.append(learn[0])
        else:
            Learning_Experience_rating.append(np.nan)

        review = re.findall(r'View all (\d+) reviews', text)
        if review:
            Total_Reviews.append(review[0])
        else:
            Total_Reviews.append(np.nan)

        fee = re.findall(r'Fees:\s*([\d\.,\sA-Za-z]+)', text)
        if fee:
            Fees.append(fee[0])
        else:
            Fees.append(np.nan)

        salary = re.findall(r'Salary\s*:\s*([\d\.,\sA-Za-z]+)', text)
        if salary:
            Salary.append(salary[0])
        else:
            Salary.append(np.nan)
            
```

```python
# Quit browser
driver.quit()
```

```python
# Create DataFrame
college_data = pd.DataFrame({
    "NIRF_Rank": NIRF_Rank,
    "Institute": Institute,
    "Placements_rating": Placements_rating,
    "Infrastructure_rating": Infrastructure_rating,
    "Faculty_Course_Curriculum_rating": Faculty_Course_Curriculum_rating,
    "Crowd_Campus_Life_rating": Crowd_Campus_Life_rating,
    "Value_for_Money_rating": Value_for_Money_rating,
    "Learning_Experience_rating": Learning_Experience_rating,
    "Total_Reviews": Total_Reviews,
    "Fees": Fees,
    "Salary": Salary
})
```

```python
# Add default column
college_data["Courses"] = "B.Tech"
```

```python
# Save to CSV
college_data.to_csv("shiksha_colleges8.csv", index=False)
print("Data saved to shiksha_colleges.csv")
```
